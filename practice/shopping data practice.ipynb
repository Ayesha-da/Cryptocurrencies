```python
df.drop(columns=["CustomerID"], inplace=True)
df
```

```python
def change_member_data(member):
    if member == "Yes":
        return 1
    else:
        return 0
    
df["Card Member"]= df["Card Member"].apply(change_member_data)
df
```

```python
df["Annual Income"]= df["Annual Income"]/1000
df
```

```python
df.columns = df.columns.str.replace(' ', '')
df
```

```python
df.columns = df.columns.str.replace('\d+', '')
df
```

```python
# saving clean data
file_path="Resources/clean_shopping_data.csv"
df.to_csv(file_path, index=False)
```
